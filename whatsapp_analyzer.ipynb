```python
import re
import pandas as pd
```

```python
# opens the file in the form of string
# file name is changed

f = open("WhatsApp Chat with Hum5.txt", "r", encoding='utf-8')
```

```python
data= f.read()
```

```python
print(data)
```

```python
#converting string dataset into pandas dataset website-regex

pattern = '\d{1,2}/\d{1,2}/\d{1,2},\s\d{1,2}:\d{2}\s(?:am|pm)\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
```

```python
messages
```

```python
dates= re.findall(pattern,data)
dates
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
df['message_date'].str.replace('pm', 'PM').str.replace('am', 'AM')  
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %I:%M %p - ')

df.rename(columns={'message_date': 'date'}, inplace =True)
df.head()
```

```python
df.shape
```

```python
# seperate user and message 

users=[]
messages=[]
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s',message)
    if entry[1:] : # user name
      users.append(entry[1])
      messages.append(entry[2])
    else:
       users.append('gruop_notification')
       messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns = ['user_message'], inplace = True)

```

```python
df
```

Extracting date, month and year from date column of the dataser

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
df['date']  = df['date'].dt.day
```
